### 4. 회귀분석

```python
# 원핫코딩(어음종류)
df_b=pd.get_dummies(df['bill'],prefix='bill')
df1=pd.concat([df,df_b],axis=1)
del df1['bill']
df1.head()
```

```python
# 원핫코딩(부도사유), local 삭제
df_r=pd.get_dummies(df['rsn'],prefix='rsn')
df2=pd.concat([df1,df_r],axis=1)
del df2['rsn']
del df2['local']
df2.head()
```

```python
#df1.to_csv("c:/Bankruptcy/data/bankruptcy_dum.csv ", index=False)
```

```python
# 독립변수 (원핫코딩 후 전체 데이터), 종속변수 (amt)
df3=pd.DataFrame(df2)
price_label = df2['amt'].copy()
df3.drop('amt',axis=1, inplace=True)
df3.head() 
```

```python
X = df3    
y= price_label
```

```python
# 회귀분석 (R-squred 0.13)
import statsmodels.api as sm

model = sm.OLS(y, X)
result = model.fit()
print("모형의적합성_R2 :",result.rsquared)
result.summary()
```

```python
#스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
df_S = scaler.fit_transform(df2) 
df_S1= pd.DataFrame(df_S,columns=df2.columns)
df_S1.boxplot()
```

```python
price_label1 = df_S1['amt'].copy()
df_S1.drop('amt',axis=1, inplace=True)
df_S1.head() 
```

```python
X = df_S1  
y= price_label1
```

```python
# 회귀분석 (R-squred 0.13)
import statsmodels.api as sm

model = sm.OLS(y, X)
result = model.fit()
print("모형의적합성_R2 :",result.rsquared)
result.summary()
```

```python
#P-value 확인
pd.options.display.float_format = '{:.5f}'.format
print(result.pvalues)
```

```python
# P-value 유효한 순으로 변수 선정
X = df_S1[{'ym','cnt', 'amt_class', 'kospi','cny','jpy','c_pi','house_pi','cnt_co', 'amt_co','bill_1','rsn_4'}] 
y= price_label
```

```python
# 변수 선정 후 회귀분석 0.12
import statsmodels.api as sm

model = sm.OLS(y, X)
result = model.fit()
result.summary()
```

```python
## VIF 산출
from statsmodels.stats.outliers_influence import variance_inflation_factor  

vif = pd.DataFrame()
vif['VIF']=[variance_inflation_factor(df3.values, i) for i in range(df3.shape[1])]
vif['독립변수'] = df3.columns
vif =vif.sort_values('VIF').reset_index(drop=True)
vif  
```

```python
# VIF 유효한 순으로 변수 선정
X = df_S1[{'ym','amt_is','cnt', 'amt_class', 'cnt_co', 'amt_py','unemp'}] 
y= price_label
```

```python
# 변수 변경 후 회귀분석 0.11
import statsmodels.api as sm

model = sm.OLS(y, X)
result = model.fit()
result.summary()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
model = LinearRegression().fit(X_train, y_train)

print("학습용 : ", model.score(X_train, y_train))
print("검증용 : ", model.score(X_test, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
model = DecisionTreeRegressor().fit(X_train, y_train)

print("학습용 : ", model.score(X_train, y_train))
print("검증용 : ", model.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
model = RandomForestRegressor(n_jobs = -1).fit(X_train, y_train)

print("학습용 : ", model.score(X_train, y_train))
print("검증용 : ", model.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

param_grid = [
 {'n_estimators': [3, 10, 30, 50, 100], 'max_features':[2, 4, 6, 8, 10, 12, 14 ]},
 ]
forest_reg = RandomForestRegressor(random_state=0,n_jobs=-1)
model = GridSearchCV(forest_reg, param_grid, cv=2)
model.fit(X_train, y_train)

print("학습용 : ", model.score(X_train, y_train))
print("검증용 : ", model.score(X_test, y_test))
```

```python
#SVR
from sklearn.svm import SVR
model = SVR(kernel="linear")
model.fit(X_train, y_train)

print("학습용 : ", model.score(X_train, y_train))
print("검증용 : ", model.score(X_test, y_test))
```

```python
#사례기반 추론 
from sklearn.neighbors import KNeighborsClassifier
#사이킨런 신경망
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
score_tr=knn.score(X_train, y_train)
score_te=knn.score(X_test, y_test)

print("학습용:",score_tr)
print("검증용:",score_te)
```
