### 1. 파일 불러오기 및 합치기


```python
# 기업부도 / 어음발행 / 어음지급 데이터 
import pandas as pd

df1= pd.read_csv("c:/bankruptcy/data/기업부도.csv")
df2= pd.read_csv("c:/bankruptcy/data/어음발행.csv")
df3= pd.read_csv("c:/bankruptcy/data/어음지급.csv")
```

```python
df1.head()
```

```python
#어음발행 데이터 2020년 8월 ~ 2022년 8월 / 다른 변수 데이터 기간 2018년 8월 ~2023년 3월과 상이함   
df2.head()
```

```python
df2.tail()
```

```python
df3.head()
```

```python
# 월별 평균으로 행 합치기
df1y = pd.DataFrame(df1.groupby('ym').mean())
df2y = pd.DataFrame(df2.groupby('ym').mean())
df3y = pd.DataFrame(df3.groupby('ym').mean())
```

```python
pd.options.display.float_format = '{:.5f}'.format
df2y.head()
```

```python
df2y.info()
```

```python
df1y=df1y.reset_index()
df2y=df2y.reset_index()
df3y=df3y.reset_index()
```

```python
dfm1=pd.merge(df1y,df3y, how="outer")
dfm2=pd.merge(dfm1,df2y, how="outer")
```

```python
#dfm2 결측값 확인 및 처리 
dfm2.isnull().sum()
```

```python
pd.options.display.float_format = '{:.5f}'.format
dfm2.fillna(dfm2.mean(), inplace=True)
dfm2.head()
```

```python
dfm2.isnull().sum()
```

```python
# 독립변수 추가 생성 / 연월평균   기업 부도 건수 및 금액, 어음발행 및 지급금액   
dfm2.head()
```

```python
# 부도어음 집계 데이터 불러오기
df4= pd.read_csv("c:/bankruptcy/data/ALL.부도어음집계.csv")
df4.head() 
```

```python
#독립변수 데이터 불러오기
df5 = pd.read_csv("c:/Bankruptcy/data/독립변수.csv")
df5.head()
```

```python
# 부도어음 + 독립변수 합치기
dfm3=pd.merge(df4,df5, how="outer")
dfm3.head()
```

```python
# 데이터셋 완성 
df=pd.merge(dfm3,dfm2, how="outer")
df.head()
```
